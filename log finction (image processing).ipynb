```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def image_log_transform(img):
    # converting the image to float to be able to use the log function
    image_float32 = img.astype(np.float32)
    # applying the log transform
    result = np.log1p(image_float32)
    # normalizing the output to the range 0-255
    result = cv2.normalize(result, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)
    return result

img = cv2.imread(r"E:\images.jpeg")

final_result = image_log_transform(img)

# Display the input and result images

# Convert BGR to RGB for proper display
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#final_result = cv2.cvtColor(final_result, cv2.COLOR_GRAY2RGB)

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(final_result)
plt.title('Final Result')
plt.axis('off')

plt.tight_layout()
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
